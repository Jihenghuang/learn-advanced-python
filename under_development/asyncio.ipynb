```python
import aiohttp
import asyncio
async with aiohttp.ClientSession() as session:
    async with session.get('https://images5.alphacoders.com/685/685514.jpg') as resp:
        with open('china.jpg', 'wb') as fd:
            pool = []
            finished = False
            while not finished:
                while len(pool)<8:
                    chunk = resp.content.read(1000)
                    print(1, chunk)
                    if not chunk:
                        break
                    pool.append(chunk)
                    
                while len(pool) > 0:
                    print(3, len(pool))
                    chunk = pool.pop(0)
                    try:
                        result = await asyncio.wait_for(chunk, timeout=0.1)
                        if result:
                            fd.write(result)
                            print(2, len(result))
                        else:
                            finished = True
                    except asyncio.TimeoutError:
                        pool.append(chunk)
                
                
                
```
